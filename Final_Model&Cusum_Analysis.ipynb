```python
import re

def extract_country_code(s):
    match = re.match(r"([a-z]+)([0-9]+)", s, re.I)
    if match:
        return match.groups()[0]
    return ''

df1['country_id'] = df1['event_id_cnty'].apply(extract_country_code)
```

```python
# List of event types
event_types = ['Protests', 'Strategic developments', 'Violence against civilians',
               'Explosions/Remote violence', 'Riots', 'Battles']

# Convert event types to columns and count occurrences per year for each country
df_combined['count'] = 1  # Helper column to count events
df_pivot = df_combined.pivot_table(index=['country', 'year'], 
                          columns='event_type', 
                          values='count', 
                          aggfunc='sum', 
                          fill_value=0).reset_index()

# Merge with the original df to add fatalities
df_merged = pd.merge(df_combined[['country', 'year', 'fatalities']], df_pivot, 
                     on=['country', 'year'], 
                     how='right')

# Now aggregate data
grouped_df = df_merged.groupby(['country', 'year']).agg({
    'fatalities': 'sum',
    'Protests': 'sum',
    'Strategic developments': 'sum',
    'Violence against civilians': 'sum',
    'Explosions/Remote violence': 'sum',
    'Riots': 'sum',
    'Battles': 'sum'
}).reset_index()

# Add a column for the sum of all event types
grouped_df['total_events'] = grouped_df[event_types].sum(axis=1)
```

```python
import numpy as np
import statsmodels.formula.api as smf
```

```python
df_final_mod['log_fatalities'] = np.log(df_final_mod['fatalities'] + 1)
df_final_mod['log_Protests'] = np.log(df_final_mod['Protests'] + 1)
df_final_mod['log_Strategic_developments'] = np.log(df_final_mod['Strategic developments'] + 1)
df_final_mod['log_Violence_against_civilians'] = np.log(df_final_mod['Violence against civilians'] + 1)
df_final_mod['log_Explosions_Remote_violence'] = np.log(df_final_mod['Explosions_Remote_violence'] + 1)
df_final_mod['log_Riots'] = np.log(df_final_mod['Riots'] + 1)
df_final_mod['log_Battles'] = np.log(df_final_mod['Battles'] + 1)
```

```python
df_final_mod
```

```python
df_final_mod['Inflation'] = pd.to_numeric(df_final_mod['Inflation'], errors='coerce')
df_final_mod['GDP'] = pd.to_numeric(df_final_mod['GDP'], errors='coerce')
df_final_mod['FDI'] = pd.to_numeric(df_final_mod['FDI'], errors='coerce')
df_final_mod['Unemployment'] = pd.to_numeric(df_final_mod['Unemployment'], errors='coerce')
```

```python
from statsmodels.formula.api import ols

# Define variables
dependent_vars = ['GDP', 'Inflation', 'Unemployment','FDI']
independent_vars = ['log_fatalities', 'log_Protests', 'log_Strategic_developments', 
                    'log_Violence_against_civilians', 'log_Explosions_Remote_violence', 
                    'log_Riots', 'log_Battles']
regions = df_final_mod['region'].unique()  # Assuming 'region' is your region column in df_final_mod

# Create empty list to store results
model_results = []

# Loop over each region and dependent variable
for region in regions:
    for dv in dependent_vars:
        # Subset data for current region
        df_region = df_final_mod[df_final_mod['region'] == region]
        
        # Define formula for regression
        formula = f"{dv} ~ {' + '.join(independent_vars)} - 1"
        
        # Fit model
        model = ols(formula=formula, data=df_region)
        result = model.fit()
        
        # Print summary
        print(f"\n\nRegion: {region}, Dependent Variable: {dv}")
        print(result.summary())
        
        # Append result to list (store region, dependent variable, and adjusted R-squared)
        model_results.append((region, dv, result.rsquared_adj))

# Create DataFrame from results
results_df = pd.DataFrame(model_results, columns=['Region', 'Dependent_Variable', 'Adj_R-squared'])

# Sort by adjusted R-squared in descending order
results_df = results_df.sort_values(by='Adj_R-squared', ascending=False)

# Print top models
print("\n\nTop models ranked by adjusted R-squared:")
print(results_df)
```

```python

from statsmodels.formula.api import ols

# Define variables
dependent_vars = ['GDP', 'Inflation', 'Unemployment', 'FDI']
independent_vars = ['log_fatalities', 'log_Protests', 'log_Strategic_developments', 
                    'log_Violence_against_civilians', 'log_Explosions_Remote_violence', 
                    'log_Riots', 'log_Battles']
countries = df_final_mod['country'].unique()  # Assuming 'country' is your country column in df_final_mod

# Create empty list to store results
model_results = []

# Loop over each country and dependent variable
for country in countries:
    for dv in dependent_vars:
        # Subset data for current country
        df_country = df_final_mod[df_final_mod['country'] == country]
        
        # Check for NaN values and sufficient data points
        if df_country.isnull().values.any() or df_country.shape[0] < 2:
            continue
        
        # Define formula for regression
        formula = f"{dv} ~ {' + '.join(independent_vars)} - 1"
        
        # Fit model
        model = ols(formula=formula, data=df_country)
        result = model.fit()
        
        # Print summary
        print(f"\n\nCountry: {country}, Dependent Variable: {dv}")
        print(result.summary())
        
        # Append result to list (store country, dependent variable, and adjusted R-squared)
        model_results.append((country, dv, result.rsquared_adj))

# Create DataFrame from results
results_df = pd.DataFrame(model_results, columns=['Country', 'Dependent_Variable', 'Adj_R-squared'])

# Sort by adjusted R-squared in descending order
results_df = results_df.sort_values(by='Adj_R-squared', ascending=False)

# Print top models
print("\n\nTop 50 models ranked by adjusted R-squared:")
print(results_df.head(50))


```
